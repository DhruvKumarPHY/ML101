Using this build a regression model and predict the per capita income fo canadian citizens in year 2020___

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv')
df
```

```python
df.head(0)
```

```python
plt.scatter(df['per capita income (US$)'],df.year, marker='+',color='red')
```

```python
new_df = df.drop('per capita income (US$)',axis='columns')
new_df
```

```python
per_capita_income = df['per capita income (US$)']
per_capita_income 
```

```python
# lets try out the linear model on it.
reg = linear_model.LinearRegression()
reg.fit(new_df,per_capita_income)
```

```python
reg.predict([[2020]])
```

```python
reg.predict([[2022]])
```
