# Binary Tree for a Salary Classification of MORE THAN 100K Dollors

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = pd.read_csv('/Users/dhruvkumar/Desktop/salaries.csv')
```

```python
df1
```

```python
inputs = df1.drop('salary_more_then_100k',axis='columns')    # we will drop the salary column as we hav eto set it as our target parameter.
target = df1['salary_more_then_100k']
```

```python
inputs
```

```python
target
```

```python
# As we know only Numerical alteration can be used in our dataset so we have to transform it accordingly.
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['degree_n'] = le_degree.fit_transform(inputs['degree'])
```

```python
inputs   # Now we will drop the Alphabatic Columns
```

```python
inputs_n = inputs.drop(['company','job','degree'],axis='columns')
```

```python
inputs_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.predict([[2,1,0]])   # Google , computer programmer , bachelors  # NO
```

```python
model.predict([[2,1,1]])  # Google , computer programmer , Masters  # Yes
```

```python
model.score(inputs_n, target)   # Perfect .......
```
