```python
cross_val_score(SVC(kernel='linear',C=15,gamma='auto'),iris1.data, iris1.target, cv=5)
```

```python
cross_val_score(SVC(kernel='linear',C=30,gamma='auto'),iris1.data, iris1.target, cv=5)
```

___Above approach is tiresome and very manual. We can use for loop as an alternative___

```python
kernels = ['rbf', 'linear']
C = [1,10,15,20,30]
avg_scores = {}
for kval in kernels:
    for cval in C:
        cv_scores = cross_val_score(SVC(kernel=kval,C=cval,gamma='auto'),iris1.data, iris1.target, cv=5)
        avg_scores[kval + '_' + str(cval)] = np.average(cv_scores)

avg_scores
#From results we can say that rbf with C=1 or 10 or linear with C=1 will give best performance
```

# Use GridSearchCV
__GridSearchCV does exactly same thing as for loop above but in a single line of code__

```python
from sklearn.model_selection import GridSearchCV
```

```python
fm = GridSearchCV(SVC(gamma='auto'),{
    
    'C' : [10,15,20,25,30],
    'kernel' : ['linear','rbf']
} , cv = 5)
fm.fit(iris1.data,iris1.target)
```

```python
fm.cv_results_
```

```python
# We will convert it to pandas dataFrame to make it easier to understand
fdf = pd.DataFrame(fm.cv_results_)
fdf.head()
```

```python
fdf[['param_C','param_kernel','mean_test_score']]
```

```python
dir(fm)
```

```python
fm.best_params_
```

```python
fm.best_score_
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rs = RandomizedSearchCV(SVC(gamma='auto'), {
        'C': [1,10,20],
        'kernel': ['rbf','linear']
    }, 
    cv=5, 
    return_train_score=False, 
    n_iter=2
)
rs.fit(iris1.data, iris1.target)
pd.DataFrame(rs.cv_results_)[['param_C','param_kernel','mean_test_score']]
# here our data will randomely run the data set for two values and show as like this.
# hence, for bigger data sets this method turn out to be game changer.
```

## How about different models with different hyperparameters?

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params' : {
            'C': [1,10,20],
            'kernel': ['rbf','linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'n_estimators': [1,5,10]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params': {
            'C': [1,5,10]
        }
    }
}
```

```python
scores = []

for model_name, mp in model_params.items():
    fm =  GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    fm.fit(iris1.data, iris1.target)
    scores.append({
        'model': model_name,
        'best_score': fm.best_score_,
        'best_params': fm.best_params_
    })
    
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```
